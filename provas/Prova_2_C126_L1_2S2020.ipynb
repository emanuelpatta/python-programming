Usar cláusulas `if` e `else` em funções lambda é bem simples, a sintaxe é a seguinte:

```python
lambda <argumentos>: <Valor de retorno se a condição é verdadeira> if <condição> else <Valor de retorno se a condição é falsa>
```
**Exemplo**: Criando uma função lambda para verificar se o valor dado está entre 10 e 20:

```python
lambda x : True if (x > 10 and x < 20) else False
```
